## Introduction

This notebook explores the **PhiUSIIL Phishing URL Dataset**, a large-scale collection of both legitimate and phishing URLs. In total, the dataset contains **134,850 legitimate** and **100,945 phishing** URLs, most of which are **recently gathered**. 

To facilitate phishing detection, we extracted various features directly from each URL and its corresponding webpage source code. In addition to these base features, we derived more advanced metrics, such as:

- **CharContinuationRate**  
- **URLTitleMatchScore**  
- **URLCharProb**  
- **TLDLegitimateProb**

These engineered features aim to capture subtle patterns in URL structure and webpage content that are often indicative of malicious intent. Throughout this notebook, we will perform data exploration, preprocessing, model training, and evaluation steps to demonstrate how these features can be leveraged to **accurately distinguish between phishing and legitimate URLs**.

# Prediction Model

## Data Preparation

Before building the machine learning model, we need to properly prepare the data:"

1) ***Feature Selection*** – Not all categorical features are necessary; some are redundant, while others do not add significant value.
2) ***Encoding*** – Categorical data should be appropriately encoded for use in the model.
3) ***Scaling*** – Numerical features need to be scaled to ensure consistency across different ranges.

### Feature Selection

```python
X_cat.head()
```

```python
# We have 695 unique TLDs
X_cat.TLD.value_counts()
```

```python
# We drop the url and domain and only keep the TLD. We also dont need the title
X_cat_reduced = X_cat.drop(['URL','Domain', 'Title', 'TLD'], axis = 1)
```

```python
X_cat_reduced.head()
```

```python
cats_features = X_cat_reduced.columns
nums_features = X_num.columns
```

### Train-Eval-Test split

```python
from sklearn.model_selection import train_test_split
#we create val and test datasets in case we want to use Deep Learning models
X_train, X_, y_train, y_ = train_test_split(X,y, test_size = 0.5, shuffle = True, stratify = y)
X_val, X_test, y_val, y_test = train_test_split(X_, y_, test_size = 0.5, shuffle = True, stratify = y_)
```

### Data Transformation Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
pipe = ColumnTransformer([
    ("Encoder", OneHotEncoder(),cats_features),
    ("Scaler", StandardScaler(), nums_features)
])
```

```python
X_train_trans = pipe.fit_transform(X_train)
X_val_trans = pipe.transform(X_val)
X_test_trans = pipe.transform(X_test)
```

```python
X_train_trans.mean(axis = 0)
```

```python
X_train_trans.std(axis = 0)
```

```python
y_train = y_train.values.reshape(-1,)
y_val = y_val.values.reshape(-1,)
y_test = y_test.values.reshape(-1,)
```

## Model Selection 

### Classical Machine Learning Models

```python
# Tree-based models
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# External Boosting Libraries
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
#rf = RandomForestClassifier(n_estimators=100, random_state=42)
xgb = XGBClassifier(n_estimators=100, eval_metric='logloss')
lgbm = LGBMClassifier(n_estimators=100, verbose = -1)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#scores_rf = cross_val_score(rf, X_train_trans, y_train, cv = 5, scoring='accuracy')
scores_xgb = cross_val_score(xgb, X_train_trans, y_train, cv = 5, scoring='accuracy')
scores_lgbm = cross_val_score(lgbm, X_train_trans, y_train, cv = 5, scoring='accuracy')
```

```python
scores_xgb.mean()
```

```python
scores_lgbm.mean()
```

OK it seems that these models perform pretty the same, lets apply train model and then evaluate it

### Train & Eval LGBM

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
lgbm.fit(X_train_trans, y_train)
```

```python
y_val_pred = lgbm.predict(X_val_trans)
```

```python
accuracy_score(y_val, y_val_pred)
```

```python
y_test_pred = lgbm.predict(X_test_trans)
```

```python
accuracy_score(y_test, y_test_pred)
```
